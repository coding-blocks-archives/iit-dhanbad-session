```
import matplotlib.pyplot as plt
```

```
img = plt.imread('prateek.jpg')
```

```
print(img.shape)
print(type(img))
```

```
plt.imshow(img)
plt.show()
```

```
combined = np.hstack((img,img,img))
border = np.zeros((60,combined.shape[1],3),dtype='uint8')+255
final = np.vstack((combined,border,combined))
plt.imshow(final)
```
